```python
import random
lit = []
for i in range(0,999):
  lit.append(str(random.randint(-1000,1000)))
print(lit)
print("hay nhap ten tep tin ")
nba = str(input())
f = open(nba, 'w+')
with open(nba, 'w') as f:
    f.writelines(lit)
f.close()
```
