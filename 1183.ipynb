```python
import random
lit = []
for i in range(0,999):
  lit.append(str(random.randint(-1000,1000)))
print("hay nhap ten tep tin ")
nba = str(input())
f = open(nba, 'w+')
with open(nba, 'w') as f:
    for i in range(len(lit)):
        if i % 10 == 0 and i > 0:
            f.writelines("\n")
            f.writelines(lit[i])
            f.writelines(", ")
        else:
            f.writelines(lit[i])
            f.writelines(", ")
f.close()
with open(nba, 'r') as f:
    z = f.read()
f.close()
z =  z.replace(",", " ")
print(z)
```
